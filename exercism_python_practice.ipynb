```python
def adj_to_verb(sentence:str,index:int):
    split_sentence = sentence.split(' ')
    if '.' in split_sentence[index]:
        remove_dot=split_sentence[index]
        return remove_dot[:-1]+"en"
    return split_sentence[index]+'en'
```

```python
def create_inventory(items: list): 
    inventory={}
    add_items(inventory,items)
    return inventory

def add_items(inventory: dict, items: list):
    for item in items:
        inventory[item]= inventory.get(item,0) + 1
    return inventory

def decrememnt_items(inventory: dict, items: list):
    for item in items:
        if item in inventory: 
            if inventory[item] > 0:
                inventory[item]-= 1
    return inventory

def remove_item(inventory: dict, item: list):
    for key in inventory:
        if key is item:
            inventory.pop(key)
            break
    return inventory

    
```

```python
def add_item(current_cart, items_to_add):
    for item in items_to_add:
        current_cart[item]= current_cart.setdefault(item,0) + 1
    return current_cart

def read_notes(notes):
    shopping_cart=dict.fromkeys(notes,1)
    return shopping_cart

def update_recipes(ideas, recipe_updates):
    ideas |=  recipe_updates
    return ideas

def sort_entries(cart):
    return dict(sorted(cart.items()))

def send_to_store(cart, aisle_mapping):
    fulfillment_cart={}
    for key, value in cart.items():
        aisle, refegeration = aisle_mapping[key]
        fulfillment_cart[key]=[value,aisle,refegeration]
    return dict(sorted(fulfillment_cart.items(),reverse=True))

def update_store_inventory(fulfillment_cart,store_inventory):
    for item, details in fulfillment_cart.items():
        ordered_qty = details[0]
        store_qty = store_inventory[item][0]
        new_qty = store_qty - ordered_qty
        if new_qty == 0:
            store_inventory[item][0]= "out of Stock"
        else:
            store_inventory[item][0]= new_qty 
    return store_inventory

```

```python
def generate_seat_letters(number):
    seat_row_letters=['A','B','C','D']
    count = 0
    while count < number:
        yield seat_row_letters[count%4]
        count +=1

def generate_seats(number):
    count=0
    while 4*count<number:
        for letter in generate_seat_letters(number):
            if letter == 'A':
                count +=1
            if count == 13:
                count += 1
            yield str(count)+letter
        
def assign_seats(passengers):
    seats_dict = dict(zip(passengers, generate_seats(len(passengers))))
    return seats_dict

def generate_codes(seat_numbers, flight_id):
    for seats in seat_numbers:
        sno_fid = seats+flight_id
        yield sno_fid+'0'*(12-len(sno_fid))

```

```python
seats = generate_seats(60)
for seat in seats:
    print(seat)
```

```python
def is_phone_number(text):
    if len(text) != 10:
        return False
    for i in range(10):
        if not text[i].isdecimal():
            return False
    return True

```

```python
message = input("enter your paragraph here: ")
for i in range(len(message)):
    segment = message[i:i+10]
    if is_phone_number(segment):
        print('Phone number found: '+ segment)
print('Done')
        
```
